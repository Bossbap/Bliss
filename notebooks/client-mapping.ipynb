```python
import pandas as pd

BASE = "/Data/baptiste.geisenberger/Data"
files = {
    "openImg":       f"{BASE}/openImg/client_data_mapping_og/train.csv",
    "google_speech": f"{BASE}/google_speech/client_data_mapping_og/train.csv",
    "femnist": f"{BASE}/femnist/client_data_mapping_og/train.csv",
    "reddit": f"{BASE}/reddit/client_data_mapping_og/train.csv"
}
```

```python
dfs = {}
for name, path in files.items():
    dfs[name] = pd.read_csv(path)
print("✅ All files loaded, keys:", list(dfs.keys()))
```

```python
# Select the dataframe
google_speech_df = dfs["google_speech"]

# Filter for client_id 1134
client_data = google_speech_df[google_speech_df["client_id"] == 1134]

# Print the datapoints
print(len(client_data))
```

```python
for name, df in dfs.items():
    print(f"▶ {name} — {len(df):,} rows × {df.shape[1]} cols")
    print("  Columns:", df.columns.tolist(), "\n")
```

```python
for name, df in dfs.items():
    print(f"▶ {name} info:")
    df.info()
    print()
```

```python
for name, df in dfs.items():
    print(f"▶ {name} — 10 random rows")
    display(df.sample(10))
```

```python
for name, df in dfs.items():
    print(f"▶ {name} — describe()")
    display(df.describe(include="all"))
```

```python
for name, df in dfs.items():
    counts = df["client_id"].value_counts().sort_index()
    print(f"▶ {name}: {counts.size:,} unique clients")
    print("  Samples per client (first 10):")
    print(counts.head(10).to_dict())
    print("  Stats:", counts.describe().to_dict(), "\n")
```
