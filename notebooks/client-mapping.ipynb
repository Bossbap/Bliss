```python
# Inspect FedScale client mapping CSV comprehensively

import os
import pandas as pd
import numpy as np

# Path to mapping CSV
CSV_PATH = "/Data/baptiste.geisenberger/Data/stackoverflow/client_data_mapping/result.csv"

# Display options to avoid truncation
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 160)
pd.set_option('display.max_colwidth', 120)
pd.set_option('display.max_rows', 50)

def fmt_bytes(n):
    try:
        n = int(n)
    except Exception:
        return str(n)
    for u in ['B','KB','MB','GB','TB','PB']:
        if n < 1024:
            return f'{n:.1f} {u}'
        n /= 1024
    return f'{n:.1f} EB'

print("=== File Info ===")
print(f"File: {CSV_PATH}")
exists = os.path.exists(CSV_PATH)
print(f"Exists: {exists}")
if not exists:
    raise FileNotFoundError(CSV_PATH)
print(f"Size : {fmt_bytes(os.path.getsize(CSV_PATH))}")

# Load CSV (let pandas infer dtypes to keep general)
df = pd.read_csv(CSV_PATH)
n_rows, n_cols = df.shape
print("\n=== DataFrame Shape ===")
print(f"Rows: {n_rows:,}, Cols: {n_cols}")

print("\n=== Columns ===")
print(list(df.columns))

print("\n=== DTypes ===")
print(df.dtypes)

try:
    df_mem = df.memory_usage(deep=True).sum()
    print(f"\nEstimated DataFrame RAM: {fmt_bytes(df_mem)}")
except Exception:
    pass

# Missing values summary
print("\n=== Missing Values Per Column ===")
nulls = df.isna().sum().to_frame('null_count')
nulls['non_null'] = len(df) - nulls['null_count']
nulls['null_frac'] = (nulls['null_count'] / len(df)).round(6)
print(nulls)

# Per-column statistics and examples
print("\n=== Per-Column Statistics and Examples ===")
for col in df.columns:
    s = df[col]
    print("\n" + "="*100)
    print(f"Column: {col}")
    print(f"  dtype: {s.dtype}")
    nuniq = s.nunique(dropna=True)
    print(f"  unique values: {nuniq:,}")
    # Numeric vs non-numeric summary
    if np.issubdtype(s.dtype, np.number):
        desc = s.describe(percentiles=[.01, .05, .25, .5, .75, .95, .99])
        print("  numeric summary:")
        print(desc)
    else:
        # Show up to 10 distinct example values
        ex_vals = s.dropna().unique()[:10]
        print("  example values (up to 10):", ex_vals)
    # Top value counts
    vc = s.value_counts(dropna=True).head(10)
    print("  top value_counts (top 10):")
    print(vc)

# Client distribution summary (if client_id present)
if 'client_id' in df.columns:
    print("\n" + "*" * 100)
    print("Client sample counts summary (by client_id)")
    cid_counts = df['client_id'].value_counts().sort_index()
    print("\nCounts describe (clients as rows):")
    print(cid_counts.describe(percentiles=[.01,.05,.25,.5,.75,.95,.99]))
    print("\nTop 10 clients by number of samples:")
    print(cid_counts.sort_values(ascending=False).head(10))
    print("\nBottom 10 clients by number of samples:")
    print(cid_counts.sort_values(ascending=True).head(10))
    # Simple imbalance metric: Gini coefficient
    def gini(x):
        v = np.sort(x.values.astype(np.float64))
        if v.size == 0:
            return np.nan
        i = np.arange(1, v.size + 1)
        return (np.sum((2*i - v.size - 1) * v) / (v.size * np.sum(v)))
    try:
        print(f"\nGini coefficient (client sample imbalance): {gini(cid_counts):.4f}")
    except Exception as e:
        print("Could not compute Gini:", e)

# sample_path integrity checks (if present)
if 'sample_path' in df.columns and np.issubdtype(df['sample_path'].dtype, np.number):
    print("\n" + "*" * 100)
    print("sample_path integrity checks")
    sp = df['sample_path']
    sp_min, sp_max, sp_n, sp_uniq = int(sp.min()), int(sp.max()), len(sp), sp.nunique()
    dups = int(sp.duplicated().sum())
    print(f"min={sp_min}, max={sp_max}, n={sp_n:,}, unique={sp_uniq:,}, duplicated_entries={dups:,}")
    if dups == 0 and sp_min == 0 and sp_max == sp_n - 1:
        print("Appears to be a contiguous 0..N-1 index (no duplicates).")
    else:
        # Show a few missing indices if close to contiguous
        try:
            if sp_max - sp_min <= 2_000_000 and sp_uniq <= 2_000_000:
                missing = sorted(set(range(sp_min, sp_max + 1)) - set(sp.astype(int).values))
                print(f"Missing indices within [{sp_min}..{sp_max}]: {len(missing)} (showing up to 20): {missing[:20]}")
            else:
                print("Range is large; skipping explicit missing-index enumeration.")
        except Exception as e:
            print("Could not enumerate missing indices:", e)

# Label-like columns quick checks
for cand in ['label_name', 'label_id']:
    if cand in df.columns:
        print("\n" + "*" * 100)
        print(f"{cand} quick summary")
        s = df[cand]
        print("value_counts (top 10, including NaN):")
        print(s.value_counts(dropna=False).head(10))
        if np.issubdtype(s.dtype, np.number):
            try:
                print("min/max:", s.min(), s.max())
            except Exception:
                pass

# Example rows
print("\n" + "="*100)
print("Examples: head(5)")
print(df.head(5))
print("\nExamples: tail(5)")
print(df.tail(5))
try:
    print("\nExamples: random sample(5)")
    print(df.sample(5, random_state=42))
except Exception:
    pass

print("\n=== Done ===")
```
